```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
#creating datasets such as standardization is not required
from sklearn.datasets import make_classification
X,Y=make_classification(n_samples=100,n_features=10,n_classes=2,random_state=42)
X = pd.DataFrame(X)
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.30,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
l = LogisticRegression()
l.fit(X_train,Y_train)
y_pred = l.predict(X_test)
y_pred
```

```python
l.predict_proba(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
s = accuracy_score(y_pred,Y_test)
c = confusion_matrix(y_pred,Y_test)
r = classification_report(y_pred,Y_test)
print(s)
print(c)
print(r)
```

HYPERPARAMETER TUNING AND CROSS VALIDATION

```python
model = LogisticRegression()
penalty = ['l1','l2','elasticnet']
c_values = [100,10,1.0,0.1,0.01]
solver = ['newton-cg','lbfgs','liblinear','sag','saga']
params = dict(penalty=penalty,C=c_values,solver=solver)
```

```python
##Grid Search CV
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
grid = GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid.fit(X_train,Y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
y_pred2 = grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
s2 = accuracy_score(y_pred2,Y_test)
c2 = confusion_matrix(y_pred2,Y_test)
r2 = classification_report(y_pred2,Y_test)
print(s2)
print(c2)
print(r2)
```

```python
from sklearn.model_selection import RandomizedSearchCV
model = LogisticRegression()
x = RandomizedSearchCV(estimator=model,param_distributions=params,scoring='accuracy_score',cv=5)
x.fit(X_train,Y_train)
```
